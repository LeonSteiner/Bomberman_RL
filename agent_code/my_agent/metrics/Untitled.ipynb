```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
losses = []
for i in range(30):
    dummy = np.load('loss{}.npy'.format(i+1))
    losses.append(np.mean(dummy))
```

```python
plt.figure(dpi=150)
plt.plot(losses)
#plt.yscale('log')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
losses = []
losses2 = []
Qs = []
for i in range(1,81//2):
    dummy = np.load('loss{}.npy'.format(i*15))
    dummyQ = np.load('Q{}.npy'.format(i*15))
    losses.extend([np.mean(dummy[j: j+100]) for j in range(len(dummy)//100)])
    losses2.extend(dummy)
    Qs.extend([np.mean(dummyQ[j: j+100]) for j in range(len(dummyQ)//100)])
    #print()
```

```python
fig, axs = plt.subplots(2, 1, dpi=150)
axs[0].plot(losses)
axs[0].set_yscale('log')
axs[1].plot(Qs)
```
